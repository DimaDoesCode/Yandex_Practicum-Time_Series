```python
def my_make_features(data, max_lag, rolling_mean_size):
    df = data.copy()
#    df['year'] = df.index.year
#    df['month'] = df.index.month
#    df['day'] = df.index.day
    df['dayofweek'] = df.index.dayofweek
    df['hour'] = df.index.hour
    
    for lag in range(1, max_lag + 1):
        df[f'lag_{lag}'] = df['num_orders'].shift(lag)

    df['rolling_mean'] = df['num_orders'].shift().rolling(rolling_mean_size).mean() 
    return df

def my_dickey_fuller_test(time_series):
    result = adfuller(time_series)
    print(f'ADF Statistic:{result[0]:.2f}')
    print(f'p-value: {result[1]:.2f}')
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'{key}: {value:.2f}')
```

### LinearRegression и поиск оптимального количества признаков.

```python
param_grid = {
    'max_lag': range(40, 60, 2),
    'rolling_mean_size': range(10, 30, 2)
}

best_rmse_score = 48
best_params = {}

for params in product(param_grid['max_lag'], param_grid['rolling_mean_size']):

    max_lag, rolling_mean_size = params
    
    taxi_featured = my_make_features(
        data=taxi_resampled,
        max_lag=max_lag,
        rolling_mean_size=rolling_mean_size
    )

    train, test = train_test_split(
        taxi_featured,
        shuffle=False,
        test_size=0.1,
        random_state=RANDOM_STATE
    ) 
    train = train.dropna()

    X_train = train.drop('num_orders', axis=1)
    y_train = train['num_orders']

    X_test = test.drop('num_orders', axis=1)
    y_test = test['num_orders']

    model_lr_features = LinearRegression()
        
    rmse_scores = cross_val_score(
        model_lr_features,
        X_train,
        y_train, 
        cv=5,
        scoring=make_scorer(mean_squared_error, squared=False, greater_is_better=True)
    )
    mean_rmse = np.mean(rmse_scores)

    if mean_rmse < best_rmse_score:
#        print(mean_rmse)
        best_rmse_score = mean_rmse
        best_params = dict(zip(param_grid.keys(), params))
        
        model_lr = model_lr_features
        
        X_train_best_featured = X_train
        y_train_best_featured = y_train

        X_test_best_featured = X_test
        y_test_best_featured = y_test
```

```python
model_lr.fit(X_train_best_featured, y_train_best_featured)
```

```python
print(f'Лучшие параметры инжиниринга признаков: {best_params}')
print(f'Лучшее RSME: {best_rmse_score:.2f}')
```

### Long short-term memory.

```python
model_lstm = Sequential([
    LSTM(50, input_shape=(X_train_best_featured.shape[1], 1)),
    Dense(1)
])

model_lstm.compile(loss='mean_squared_error', optimizer='adam', metrics=[RootMeanSquaredError()])

tss = TimeSeriesSplit(n_splits=5)

rmse_scores = []

scaler_X= MinMaxScaler()
scaler_y = MinMaxScaler()


for train_index, test_index in tss.split(X_train_best_featured):
    X_train_tss, X_test_tss = X_train_best_featured.iloc[train_index], X_train_best_featured.iloc[test_index]
    y_train_tss, y_test_tss = y_train_best_featured.iloc[train_index], y_train_best_featured.iloc[test_index]

    X_train_scaled = scaler_X.fit_transform(X_train_tss)
    X_test_scaled = scaler_X.transform(X_test_tss)
    
    y_train_scaled = scaler_y.fit_transform(y_train_tss.values.reshape(-1, 1))

    model_lstm.fit(X_train_scaled, y_train_scaled, epochs=5, batch_size=10, verbose=0)
    y_pred_tss_scaled = model_lstm.predict(X_test_scaled)

    y_pred_tss = scaler_y.inverse_transform(y_pred_tss_scaled)

    rmse_tss = np.sqrt(mean_squared_error(y_test_tss, y_pred_tss))
    rmse_scores.append(rmse_tss)
```

```python
avg_rmse = np.mean(rmse_scores)
print(f'Average RMSE on cross-validation: {avg_rmse:.2f}')
```

```python
model_lstm.fit(
    scaler_X.transform(X_train_best_featured),
    scaler_y.transform(y_train_best_featured.values.reshape(-1, 1)),
    epochs=100,
    batch_size=10,
    verbose=0
)
```
